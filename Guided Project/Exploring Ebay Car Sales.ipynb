### Based on the information, we do need some data cleaning
- some variables have mixed strings and numbers
- dates are in string format
- name variable needs to separate the brand, model, & number
- probably, some have null values
- column names use [camelcase](https://en.wikipedia.org/wiki/Camel_case) instead of Python's preferred [snakecase](https://en.wikipedia.org/wiki/Snake_case)

```python
# Let's clean the columns first
# Make a map dictionary for the ff columns
column_dict = {'yearOfRegistration': 'registration_year',
                'monthOfRegistration': 'registration_month',
                'notRepairedDamage': 'unrepaired_damage',
                'dateCreated': 'ad_created'
}
# Built-in rename function to rename columns
autos.rename(columns=column_dict, inplace=True)
# Make new variable for columns list and modified back to the dataframe
column_names = list(autos.columns)
# Make a function to change camelcase to snakecase
def change_case_string(name):
    new_name = ""
    for character in name:
        if character.isupper():
            # if found, change to lower and add underscore char
            character = "_" + character.lower()
            new_name += character
        else:
            new_name += character
    return new_name

def clean_list_case(lists):
    # Made list comprehension for one line code
#     new_lists = []
#     for item in lists:
#         new_lists.append(change_case_string(item))    
    return [change_case_string(item) for item in lists]

autos.columns = clean_list_case(column_names)
```

### Price and Odometer have mixed string and number. Do some data cleaning. Columns that perharps could be dropped
- seller
- offer type
- unrepaired damage
- nr of pictures

```python
# Check unique records for price and odometer
print(autos['price'].unique(), len(autos['price'].unique()))
```

```python
print(autos['odometer'].unique(), len(autos['odometer'].unique()))
```

The strings of price column are '$' and ','.
While strings of odometer are ',', 'km' and 'k'.

```python
# Cleaning price and odometer records
autos['price'] = autos['price'].str.replace('$','')
autos['price'] = autos['price'].str.replace(',','')
# Convert to numberic dtype
autos['price'] = autos['price'].astype(int)
# for odometer
autos['odometer'] = autos['odometer'].str.replace(',','')
autos['odometer'] = autos['odometer'].str.replace('km','')
autos['odometer'] = autos['odometer'].str.replace('k','')
autos['odometer'] = autos['odometer'].astype(int)
```

### Let's look at the brand column
- select the top 20 or total values > 5%
- make a dictionary to hold the aggregate data with brand name as the key and mean price is the value
- sort and display by highest mean price

```python
autos['brand'].value_counts().index
```

```python
autos['brand'].value_counts()
```

```python
# Storing the brand and mean price into a dictionary
top_brand_cars = {}
brands = autos['brand'].unique()
for b in brands:
    selected_rows = autos[autos['brand'] == b]
    price = selected_rows['price'].mean()
    if b not in top_brand_cars:
        top_brand_cars[b] = price
```

```python
sorted(top_brand_cars.items(), key=lambda kv: kv[1], reverse=True)
```

```python
# Codes from the solution
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)

brand_mean_prices = {}
for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

```python
sorted(brand_mean_prices.items(), key=lambda kv: kv[1], reverse=True)
```

```python
# Top brands with mean mileage
brand_mean_mileage = {}
for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_mileage = brand_only['odometer_km'].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

brand_mean_mileage
```

```python
sorted(brand_mean_mileage.items(), key=lambda kv: kv[1], reverse=True)
```

```python
# Make pandas series and dataframe constructor
bmp_series = pd.Series(brand_mean_prices)
bmm_series = pd.Series(brand_mean_mileage)
df = pd.DataFrame(bmp_series, columns=['mean price'])
df['mean_mileage'] = bmm_series
df
```

Based on sorting the highest values, the top brands mean price has the top brands mean mileage.
