# Visualizing Earnings Based on College Majors

- this project is to explore how to make plotting using matplot.
- the dataset is about studetns who graduated from college and outcome for jobs

Using visualizations, we'l have questions from the dataset like:
- Do studing in more popular majors make more money? using scatter plots
- How many majors are predomianntly male? Predominantly female? using historgrams
- Which category of majors have the most students? using bar plots

Information about the columns in the datasets:

    Rank - Rank by median earnings (the dataset is ordered by this column).
    Major_code - Major code.
    Major - Major description.
    Major_category - Category of major.
    Total - Total number of people with major.
    Sample_size - Sample size (unweighted) of full-time.
    Men - Male graduates.
    Women - Female graduates.
    ShareWomen - Women as share of total.
    Employed - Number employed.
    Median - Median salary of full-time, year-round workers.
    Low_wage_jobs - Number in low-wage service jobs.
    Full_time - Number employed 35 hours or more.
    Part_time - Number employed less than 35 hours.

```python
# Assign a variable to the number of rows
raw_data_count = len(recent_grads)
# Drop the missing values
recent_grads.dropna(axis=0, inplace=True)
# Assgin new variable after the drop
clean_data_count = len(recent_grads)

print(raw_data_count)
print(clean_data_count)
```
